```python
import pandas as pd
import matplotlib.pyplot as plt
import h2o
from h2o.estimators import H2ODeepLearningEstimator
import numpy as np
import graphviz
from sklearn.model_selection import train_test_split
import random
from datetime import datetime

h2o.init()

train = pd.read_csv("C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\train.csv")
test = pd.read_csv("C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\test.csv")
sample = pd.read_csv("C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\sample.csv")
    
path = "C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\"

sample_df = h2o.H2OFrame(sample)
sample_ids = sample.iloc[:, 0].to_list()

# Default value is 0.2
# Value 0.2 is for presentation purposes
TestSize = 0.02

# Default value is None
# Value 0.2 is for presentation purposes
TrainSize = 0.02

train, validation = train_test_split(train, test_size=TestSize, train_size=TrainSize)

train.hex = h2o.H2OFrame(train, destination_frame='train')
test.hex = h2o.H2OFrame(test, destination_frame='test')
validation.hex = h2o.H2OFrame(validation, destination_frame='validation')

train.hex['label'] = train.hex['label'].asfactor()

input_parameters = {
    "hidden": [[100, 100], [125, 125], [150, 150]],
    "epochs": [2, 2, 2],
    "input_dropout_ratio": [0.1, 0.2, 0.3],
    "hidden_dropout_ratios": [[0.25, 0.1], [0.5, 0.1], [0.1, 0.2]],
    "nfolds": [2, 2, 2]
}

# Genetic Algorithm parameters
population_size = 2
mutation_rate = 0.1
generations = 2

# Global Variables (not alter)
#curr_gen = 0
curr_time = 0

# Crossover parameters
uniform_crossover_rate = 0.5
# Represents the point where to break (50% -> Middle).
one_point_crossover_rate = 0.5
# Multi point MIN should be < than Multi point MAX.
# Represents the interval of value to replace (25%-75% -> first quarter until last quarter).
multi_point_min_crossover_points_rate = 0.25
multi_point_max_crossover_points_rate = 0.75

# Options Parameters
# Crossover types: uniform , one-point , multi-point
# Selection types: roulette_wheel , fitness_selection , tournment
crossover_type = "uniform"
selection_type = "fitness_selection"

TournmentSize = 3

# Fitness function
def fitness_function(inputs):
    model = H2ODeepLearningEstimator(
        model_id='model_DL',
        distribution="multinomial",
        hidden=inputs["hidden"],
        epochs=inputs["epochs"],
        nfolds=inputs["nfolds"],
        input_dropout_ratio=inputs["input_dropout_ratio"],
        hidden_dropout_ratios=inputs["hidden_dropout_ratios"],
        reproducible=True,
        activation="RectifierWithDropout",
        single_node_mode=False,
        balance_classes=True,
        force_load_balance=False,
        seed=23123,
        score_training_samples=0,
        score_validation_samples=0,
        keep_cross_validation_predictions=True,
        keep_cross_validation_fold_assignment=True,
        stopping_rounds=0,
        auc_type="macroovr"
    )

    model.train(x=list(train.columns[2:]), y=train.columns[1], training_frame=train.hex, validation_frame=validation.hex)
    
    global ModelToPlot
    ModelToPlot = model
    
    # Predict on the test data
    predicts = model.predict(test_data=test.hex).as_data_frame()["predict"].astype(int)
    predicts_auc = model.predict(test_data=train.hex).as_data_frame()["predict"].astype(int)
    predicts_ = predicts.to_list()
    
    results = pd.DataFrame({"id": sample_ids, "label": predicts_})
    
    curr_dt = datetime.now()
    timestamp = str(int(round(curr_dt.timestamp())))
    timestamp_f = timestamp + ".csv"
    
    path_f = path + timestamp_f
    results.to_csv(path_f, index=False)
    
    accuracy = (predicts_auc == train.hex['label'].as_data_frame()['label']).mean()
    print("teste 1 - "+ str(accuracy) +" - Gen: " + str(curr_gen))
    
    file = open(path+timestamp+".txt","a")
    file.write("Generation: " + str(curr_gen) + "\nAccuracy: " + str(accuracy) + "\nHidden: " + str(inputs["hidden"]) + "\nEpochs: " + str(inputs["epochs"]) + "\nNfolds: " + str(inputs["nfolds"]) + "\nInput_dropout_ratio: " + 
               str(inputs["input_dropout_ratio"]) + "\nHidden_dropout_ratios: " + str(inputs["hidden_dropout_ratios"]))
    file.close()
    
    global curr_time
    curr_time = timestamp
    return accuracy

# Perform uniform crossover
def uniform_crossover(parent1, parent2):
    child = {}
    for key in input_parameters.keys():
        if random.random() < uniform_crossover_rate:
            child[key] = parent1[key]
        else:
            child[key] = parent2[key]
    return child

# Perform one-point crossover
def one_point_crossover(parent1, parent2):
    child = {}
    crossover_point = random.randint(1, len(input_parameters) - 1)
    keys = list(input_parameters.keys())
    for i, key in enumerate(keys):
        if i < crossover_point:
            child[key] = parent1[key]
        else:
            child[key] = parent2[key]
    return child

# Perform multi-point crossover
def multi_point_crossover(parent1, parent2):
    child = {}
    keys = list(input_parameters.keys())
    for key in keys:
        if random.random() < multi_point_min_crossover_points_rate:
            child[key] = parent1[key]
        elif random.random() > multi_point_max_crossover_points_rate:
            child[key] = parent2[key]
        else:
            child[key] = parent1[key] if random.random() < 0.5 else parent2[key]
    return child

# Perform mutation
def mutation(individual):
    for key in input_parameters.keys():
        if random.random() < mutation_rate:
            individual[key] = random.choice(input_parameters[key])
    return individual

# Selection methods
def roulette_wheel_selection(population):
    fitness_sum = sum(fitness_function(individual) for individual in population)
    selection_probs = [fitness_function(individual) / fitness_sum for individual in population]
    return random.choices(population, weights=selection_probs)[0]

def tournament_selection(population):
    tournament_size = TournmentSize
    tournament_candidates = random.sample(population, tournament_size)
    best_candidate = max(tournament_candidates, key=fitness_function)
    return best_candidate

def fitness_selection(population):
    sorted_population = sorted(population, key=fitness_function, reverse=True)
    parent1 = sorted_population[0]
    parent2 = sorted_population[1]
    return parent1, parent2

# Selection function
def selection(population, selection_type):
    if selection_type == "roulette_wheel":
        return roulette_wheel_selection(population)
    elif selection_type == "tournament":
        return tournament_selection(population)
    elif selection_type == "fitness_selection":
        return fitness_selection(population)
    else:
        raise ValueError("Invalid selection type")

# Genetic Algorithm
def genetic_algorithm(crossover_type, selection_type):
    population = []
    for _ in range(population_size):
        individual = {}
        for key in input_parameters.keys():
            individual[key] = random.choice(input_parameters[key])
        population.append(individual)

    best_individual = None
    best_fitness = 0.0
    fitness_history = []
    global FinalPlot
    
    for generation in range(generations):
        offspring = []
        while len(offspring) < population_size:
            if selection_type != "fitness_selection":
                parent1 = selection(population, selection_type)
                parent2 = selection(population, selection_type)
            else:
                parent1, parent2 = selection(population, selection_type)

            if crossover_type == "uniform":
                child = uniform_crossover(parent1, parent2)
            elif crossover_type == "one-point":
                child = one_point_crossover(parent1, parent2)
            elif crossover_type == "multi-point":
                child = multi_point_crossover(parent1, parent2)
            else:
                raise ValueError("Invalid crossover type")

            child = mutation(child)
            offspring.append(child)

        population = offspring

        for individual in population:
            fitness = fitness_function(individual)
            if fitness > best_fitness:
                best_individual = individual
                best_fitness = fitness
                FinalPlot = ModelToPlot
                
        global curr_gen
        curr_gen = generation + 1
        temp = [best_fitness,curr_gen]
        fitness_history.append(temp)
        print("Generation:", curr_gen)
        print("Best Fitness:", best_fitness)
        
    FinalPlot.show()
    FinalPlot = FinalPlot.scoring_history()
    # Plot fitness history
    plt.plot(FinalPlot.epochs, FinalPlot.training_rmse, c='blue')
    plt.plot(FinalPlot.epochs, FinalPlot.validation_rmse, c='orange')
    plt.title("Scoring history - RMSE")
    plt.xlabel("Epochs")
    plt.ylabel("RMSE")
    fig = plt.gcf()
    plt.show()
    fig.savefig(path+str(curr_time)+".png",dpi=100)
    
    return best_individual, best_fitness

# Run the genetic algorithm
best_individual, best_fitness = genetic_algorithm(crossover_type, selection_type)

print("Best Individual: ", best_individual)
print("Best Fitness: ", best_fitness)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import h2o
from h2o.estimators import H2ODeepLearningEstimator
import numpy as np
import graphviz
from sklearn.model_selection import train_test_split
import random

h2o.init()

train = pd.read_csv("C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\train.csv")
test = pd.read_csv("C:\\Users\\ruimi\Downloads\\lei-ia-2223-battleground\\test.csv")

train, validation = train_test_split(train,test_size=0.20)

train.hex = h2o.H2OFrame(train, destination_frame='train')
test.hex = h2o.H2OFrame(test, destination_frame='test')
validation.hex = h2o.H2OFrame(validation, destination_frame='validation')

train.hex['label'] = train.hex['label'].asfactor()
```
